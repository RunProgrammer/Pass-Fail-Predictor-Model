```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split

# LOADIND DATA  
data = pd.read_csv("Give Your data file")
x = data.drop(columns=["Select your columns"])
y = data["select columns"]
x
y
# pd.get_dummies(data) The data parameter is the data to be converted to dummy variables.
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2) #used to split for train & test 20%
```

```python
x_train.head()
```

```python
y_train.head()
```

#IMPLEMENTING THE MODEL 

```python
from sklearn.metrics import accuracy_score


model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=32,activation="relu",input_shape=(2,)),
    tf.keras.layers.Dense(units=66,activation="relu"),
    tf.keras.layers.Dense(units=1,activation="sigmoid")
])

model.compile(loss='binary_crossentropy',optimizer = 'sgd',metrics = "accuracy")



```

```python
model.fit(x_train,y_train,epochs=600,batch_size=32)
```

```python

ask_hours = int(input("Enter the number of hours : "))
ask_score = int(input("Enter the previous score : "))

data_set = (ask_hours,ask_score)
result = model.predict([data_set])
print(result)

#result = [0 if val < 0.5 else 1 for val in result]
if result < 0.5:
    result = 0
    result = "Study Well If it continues You will fail :("
else:
    result = 1
    result = "Good Keep Going You will pass"

result
#print(len(x_test))
#accuracy_score(y_test,result)


```
